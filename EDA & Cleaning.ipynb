# Data Cleaning

Removed Column that was not needed.

Checked for non-standard values in the categorical columns.

We observed during exploration stage that trees that are not alive or are stump have few information missing. So we should remove trees that are not alive.

An easier way is to remove trees that are alive.

Instead of removing missing values we can change them to 'Not Applicable' as health of a stump or a dead is not important.

```python
# creating a mask/variable containing only stump and dead trees.
mask = ((data_subset['status'] == 'Stump') | (data_subset['status'] == 'Dead'))
# mask = data_subset['status'] != 'Alive'
```

```python
data_subset.loc[mask] = data_subset.loc[mask].fillna('Not Applicable')
```

we would also need different dataset for dead/stump and alive trees as the value for both are different.

```python
# we would also need different dataset for dead/stump and alive trees as the value for both are different.

data_subset_alive = data_subset[data_subset['status'] == 'Alive']
data_subset_dead_or_stump = data_subset[(data_subset['status'] == 'Dead') | (data_subset['status'] == 'Stump')]
```

```python
# grouping the dataset by the type of tree.

data_subset_alive.groupby('spc_latin')['tree_dbh'].describe()
```

Now we have encountered a new issue.
We can see that min value is 0 which is not possible.
SO, to rectify this situation we will set boundaries. 
We only need value lower that 25%ile and higher than 75%ile.

or we can impute the 0 to mean values.

We will go with first as it is more interesting.

```python
stats = data_subset_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats
# resetting index to spc_latin
```

```python
# mergeing the new dataset(stats) with the original(data_subset_alive)

data_subset_alive = data_subset_alive.merge(stats, on= 'spc_latin', how= 'left')
```

```python
data_subset_alive.head()
```

```python
# dataframe.loc[]: Access a group of rows and columns by label(s)

# Setting the tree_dbh as the 25% or 75% if it is out of bounds accordingly

cond = data_subset_alive['tree_dbh'] < data_subset_alive['25%'] 
#if tree diameter is smaller than 25% column, add to cond.
data_subset_alive.loc[cond, 'tree_dbh'] = data_subset_alive['25%']
#if cond is true then dbh = 25% column value.

cond = data_subset_alive['tree_dbh'] > data_subset_alive['75%']
data_subset_alive.loc[cond, 'tree_dbh'] = data_subset_alive['75%']
```

```python
# Seems all the issues are resolved.
# And we finally have the dataset with the all the trees that are alive.

data_subset_alive
```
